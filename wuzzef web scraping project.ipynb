```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
title = []
typejob = []
companys = []
locations = []
```

```python
url = "https://wuzzuf.net/a/Jobs-in-Cairo"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
soup
```

```python
soup.prettify()
```

```python
soup.find_all('div',{'class':"css-1gatmva e1v1l3u10"})
```

```python
job_title= soup.find_all('h2',{'class' :"css-m604qf"})
```

```python
job_title
```

```python
job_type = soup.find_all('span', {'class' :"css-1ve4b75 eoyjyou0"})
```

```python
job_type
```

```python
company = soup.find_all('a',{'class' :"css-17s97q8"})
```

```python
company
```

```python
location =  soup.find_all('span',{'class' :"css-5wys0k"})
```

```python
location
```

```python
for i in range(len(job_title)) :
    title.append(job_title[i].text)
    typejob.append(job_type[i].text)
    companys.append(company[i].text)
    locations.append(location[i].text)
    
print(title)
print(typejob)
print(companys)
print(locations)
     
```

```python
import pandas as pd
```

```python
jobs = pd.DataFrame(data= {'title' : title, 'job_type':job_type, 'company':companys, 'locations' : locations})
```
